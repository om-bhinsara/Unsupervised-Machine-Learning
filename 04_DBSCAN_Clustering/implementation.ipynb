```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_moons

X, y = make_moons(n_samples=300, noise=0.05, random_state=42)

plt.scatter(X[:, 0], X[:, 1], c=y)
plt.title("Two Moons Dataset")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

```python
X # coordinates of points (features)
```

```python
X.shape # 300 rows with two features
```

```python
y.shape
```

```python
y # in which cluster each point belongs
```

```python
from sklearn.cluster import DBSCAN
db=DBSCAN(eps=0.2, min_samples=4)
db.fit(X)
db.labels_
```

```python
# original plot
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.title("Two Moons Dataset")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

```python
labels = db.labels_ # get cluster labels and give them as y (in which cluster they belong)

# Plot
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')

plt.title("DBSCAN Clustering")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")

plt.show()
```
